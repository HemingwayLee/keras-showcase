```python
import MeCab

mecab1 = MeCab.Tagger("-Owakati")

with open("raw.txt", "r") as f:
    lines = list(filter(None, f))

output = []
for line in lines:
    output.extend(mecab1.parse(line).split(" "))
    
print(output)
```

```python
with open('ner.csv', 'w') as f:
    for item in output:
        f.write(f"{item}\n")
```
